1. Calculadora de IMC (Índice de Masa Corporal):

Cree una función que calcule el IMC (Índice de Masa Corporal) de una persona a partir de su peso y altura. La función debe recibir como parámetros el peso (en kg) y la altura (en cm) y devolver el valor del IMC.

Opcional: Implemente una segunda función que clasifique el IMC según los estándares de la OMS:

Bajo peso: IMC < 18.5
Peso normal: 18.5 <= IMC < 25
Sobrepeso: 25 <= IMC < 30
Obesidad: IMC >= 30


```python
def indiceMasaCorporal(peso,altura):
    imc = peso/ ((altura/100) ** 2)
    return imc
def calsificacion(imc):
    if imc<18.5:
        return "Tu indice de masa corporal se considera como: Bajo Peso"
    elif imc<25:
        return "Tu indice de masa corporal se considera como: Peso Normal"
    elif imc< 30:
        return "Tu indice de masa corporal se considera como: Sobrepeso"
    else:
        return "Tu indice de masa corporal se considera como: Obesidad"

peso = float(input("Ingrese su peso en kg: "))
altura = float(input("Ingrese su altura en centimetros: "))
imc=indiceMasaCorporal(peso,altura)
print("Su Indice de Masa Corporal es: ",imc)
print("Clasificación: ",calsificacion(imc))
```

2. Analizador de números:

Cree una función que reciba un número entero como parámetro y devuelva un diccionario con la siguiente información:

Si el número es par o impar
Si el número es positivo, negativo o cero
Si el número es divisible por 3
Si el número es divisible por 5
Si el número es divisible por 3 y por 5

```python
def diccionario(num):
    resultado = {}
    resultado['par o impar'] = 'par' if num % 2 == 0 else 'impar'
    resultado['positivo o negativo'] = 'positivo' if num > 0 else 'negativo' if num < 0 else 'cero'
    resultado['divisible por 3'] = True if num % 3 == 0 else False
    resultado['divisible por 5'] = True if num % 5 == 0 else False
    resultado['divisible por 3 y 5'] = True if num % 3 == 0 and num % 5 == 0 else False
    return resultado
num = float(input("Ingrese un numero entero: "))
print(diccionario(num))
 
```

Opcional: Implemente la función utilizando una función lambda para cada una de las verificaciones.

```python
def analizar_numero(num):
    par = lambda x: x % 2 == 0
    positivo = lambda x: x > 0
    divisible_por_3 = lambda x: x % 3 == 0
    divisible_por_5 = lambda x: x % 5 == 0
    divisible_por_3_y_5 = lambda x: divisible_por_3(x) and divisible_por_5(x)
    return {
        "Es par": par(num),
        "Es positivo": positivo(num),
        "Es divisible por 3": divisible_por_3(num),
        "Es divisible por 5": divisible_por_5(num),
        "Es divisible por 3 y 5": divisible_por_3_y_5(num)
        }

num = float(input("Ingrese un numero entero: "))
resultado = analizar_numero(num)
print(f"Análisis del número {num}:")
for i, j in resultado.items():
    print(f"{i}: {j}")
```
