<u><em>Ejercicio N° 1: Sistema de Gestión de Empleados</em></u>

Imagina que estás desarrollando un sistema de gestión de empleados para una empresa. Tu tarea es implementar un sistema que permita gestionar la información de los empleados y realizar ciertas operaciones sobre ellos.

1. Crea una clase llamada Empleado que tenga los siguientes atributos:

+ nombre: una cadena que representa el nombre del empleado.
+ edad: un número entero que indica la edad del empleado.
+ salario: un número decimal que indica el salario del empleado.
+ activo: un valor booleano que indica si el empleado está activo en la empresa.

2. La clase Empleado debe tener los siguientes métodos:

+ informacion(): muestra por pantalla la información del empleado, incluyendo el nombre, edad y salario.
+ activar(): cambia el estado del empleado a activo.
+ desactivar(): cambia el estado del empleado a inactivo.

3. Crea una clase derivada llamada Gerente que herede de la clase Empleado. La clase Gerente tiene un atributo adicional:

+ departamento: una cadena que indica el departamento del que es responsable el gerente.

4. Sobrescribe el método informacion() en la clase Gerente para incluir el departamento al mostrar la información.






```python
class Empleado:
    def __init__(self,nombre,edad,salario,activo):
        self.nombre = nombre
        self.edad = edad
        self.salario = salario
        self.activo = activo
    
    def informacion(self):
        print(f'El nombre del empleado es {self.nombre}, su edad es {self.edad} y su salario es {self.salario}')
    
    def activar(self,activo):
        if activo == False:
            activo = True
    def desactivar(self,activo):
        if activo == True:
            activo = False

class Gerente(Empleado):
    def __init__(self, nombre, edad, salario, activo,departamento):
        super().__init__(nombre, edad, salario, activo)
        self.departamento = departamento
    def informacion(self):
        super().informacion()
        print(f'Su departamento es {self.departamento}')

#empleado1 = Empleado("Juan",23,450000,True)
#empleado1.informacion()
gerente1 = Gerente("Pablo",23,850000,True,"Marketing")
gerente1.informacion()
```

<u><em>Ejercicio N° 2: Sistema de Gestión de Vehículos</em></u>

Imagina que estás desarrollando un sistema de gestión de vehículos para una empresa de alquiler de coches. Tu tarea es implementar un sistema que permita gestionar la información de los vehículos y realizar ciertas operaciones sobre ellos.

1. Crea una clase llamada Vehiculo que tenga los siguientes atributos:

+ marca: una cadena que representa la marca del vehículo.
+ modelo: una cadena que indica el modelo del vehículo.
+ precio_por_dia: un número decimal que indica el precio de alquiler por día del vehículo.
+ disponible: un valor booleano que indica si el vehículo está disponible para alquiler.

2. La clase Vehiculo debe tener los siguientes métodos:

+ informacion(): muestra por pantalla la información del vehículo, incluyendo la marca, modelo y precio por día.
+ alquilar(): cambia el estado del vehículo a no disponible.
+ devolver(): cambia el estado del vehículo a disponible.

3. Crea una clase derivada llamada Coche que herede de la clase Vehiculo. La clase Coche tiene un atributo adicional:

+ tipo_combustible: una cadena que indica el tipo de combustible del coche (por ejemplo, gasolina, diésel, eléctrico).

4. Sobrescribe el método informacion() en la clase Coche para incluir el tipo de combustible al mostrar la información.


```python
class Vehiculo:
    def __init__(self,marca,modelo,precio_por_dia,disponible):
        self.marca = marca
        self.modelo = modelo
        self.precio_por_dia = precio_por_dia
        self.disponible = disponible
    
    def informacion(self):
        print(f'La marca del vehiculo es {self.marca}, el modelo es {self.modelo} y su precio por día es {self.precio_por_dia}')
    
    def alquilar(self):
        if disponible == True:
            disponible = False
    
    def devolver(self):
        if disponible == False:
            disponible = True

class Coche(Vehiculo):
    def __init__(self, marca, modelo, precio_por_dia, disponible,tipo_combustible):
        super().__init__(marca, modelo, precio_por_dia, disponible)
        self.tipo_combustible = tipo_combustible
    
    def informacion(self):
        super().informacion()
        print(f'Y el tipo de combustible que usa es {self.tipo_combustible}')

```

<u><em>Ejercicio N° 3: Sistema de Gestión de Productos</em></u>

Imagina que estás desarrollando un sistema de gestión de productos para una tienda en línea. Tu tarea es implementar un sistema que permita gestionar la información de los productos y realizar ciertas operaciones sobre ellos.

1. Crea una clase llamada Producto que tenga los siguientes atributos:

+ nombre: una cadena que representa el nombre del producto.
+ categoria: una cadena que indica la categoría del producto.
+ precio: un número decimal que indica el precio del producto.
+ en_stock: un valor booleano que indica si el producto está en stock.

2. La clase Producto debe tener los siguientes métodos:

+ informacion(): muestra por pantalla la información del producto, incluyendo el nombre, categoría y precio.
+ vender(): cambia el estado del producto a no en stock.
+ reponer(): cambia el estado del producto a en stock.

3. Crea una clase derivada llamada ProductoElectronico que herede de la clase Producto. La clase ProductoElectronico tiene un atributo adicional:

+ garantia: un número entero que indica los meses de garantía del producto electrónico.

4. Sobrescribe el método informacion() en la clase ProductoElectronico para incluir los meses de garantía al mostrar la información.

```python
class Producto():
    def __init__(self, nombre, categoria, precio, en_stock):
        self.nombre = nombre
        self.categoria = categoria
        self.precio = precio
        self.en_stock = en_stock
    
    def informacion(self):
        print (f'El {self.nombre} de la categoría {self.categoria}, su precio es {self.precio}.')
    
    def vender(self):
        self.en_stock = False
    
    def reponer(self):
        self.en_stock = True

class ProductoElectronico(Producto):
    def __init__(self, nombre, categoria, precio, en_stock, garantia):
        super().__init__(nombre, categoria, precio, en_stock)
        self.garantia = garantia

    def informacion(self):
        super().informacion()
        print (f'El producto tiene una garantía de {self.garantia} años.')

producto1 = Producto("Sillon","Mueble",250000,True)
productoElectronico1 = ProductoElectronico("Televisor","Electrodomestico",360000,True,2)
productoElectronico1.informacion()
```

<u><em>Ejercicio N° 4: Sistema de Gestión de Cursos</em></u>

Imagina que estás desarrollando un sistema de gestión de cursos para una plataforma educativa en línea. Tu tarea es implementar un sistema que permita gestionar la información de los cursos y realizar ciertas operaciones sobre ellos.

1. Crea una clase llamada Curso que tenga los siguientes atributos:

+ titulo: una cadena que representa el título del curso.
+ instructor: una cadena que indica el nombre del instructor del curso.
+ precio: un número decimal que indica el precio del curso.
+ disponible: un valor booleano que indica si el curso está disponible para inscripción.

2. La clase Curso debe tener los siguientes métodos:

+ informacion(): muestra por pantalla la información del curso, incluyendo el título, instructor y precio.
+ abrir_inscripciones(): cambia el estado del curso a disponible.
+ cerrar_inscripciones(): cambia el estado del curso a no disponible.

3. Crea una clase derivada llamada CursoOnline que herede de la clase Curso. La clase CursoOnline tiene un atributo adicional:

+ duracion: un número entero que indica la duración del curso en semanas.

4. Sobrescribe el método informacion() en la clase CursoOnline para incluir la duración al mostrar la información.

```python
class Curso:
    def __init__(self, titulo, instructor, precio, disponible):
        self.titulo = titulo
        self.instructor = instructor
        self.precio = precio
        self.disponible = disponible
    
    def informacion(self):
        print(f'El curso de {self.titulo} tiene como instructor a {self.instructor}, y tiene un precio de {self.precio}')

    def abrir_inscripciones(self):
        self.disponible == True
    
    def cerrar_inscripciones(self):
        self.disponible == False

class CursoOnline(Curso):
    def __init__(self, titulo, instructor, precio, disponible, duracion):
        super().__init__(titulo, instructor, precio, disponible)
        self.duracion = duracion
    
    def informacion(self):
        super().informacion()
        print(f'Tiene una duracion de {self.duracion}')
```

Crear un gestor de tareas que administre una lista de tareas pendientes. La clase debe tener los siguientes metodos:
1. agregar_tarea():Este metodo debe recibir una tarea y agregarla a la lista de tareas pendientes.
2. marcar_completada(tarea): Este metodo debe marcar una tarea especifica como completada en la lista de tareas pendientes.
3. listar_tareas_pendientes(): Este metodo debe devolver una lista con todas las tareas pendientes
4. listar_tareas_completadas(): Este metodo debe devolver una lista con todas las tareas completadas.

```python
lista_tareas_pendientes = []
lista_tareas_completadas = []
class GestorTareas:
    def __init__(self):
        #self.tareas = []
        pass


    def agregar_tarea():
        tarea = input("Ingrese la tarea a realizar: ")
        lista_tareas_pendientes.append(tarea)
        print("Tarea agregada con éxito")
    def marcar_completada(tarea):
        lista_tareas_pendientes.remove(tarea)
        lista_tareas_completadas.append(tarea)
    
    def listar_tareas_pendientes():
        print("Tareas pendientes:")
        for tarea in lista_tareas_pendientes:
            print(tarea)
    def listar_tareas_completadas():
        print("Tareas completadas:")
        for tarea in lista_tareas_completadas:
            print(tarea)


    print("1. Agregar tarea")
    print("2. Marcar tarea como completada")
    print("3. Listar tareas pendientes")
    print("4. Listar tareas completadas")
    print("5. Salir")
while True:
    opcion = input("Ingrese la opción deseada: ")
    if opcion == "1":
        GestorTareas.agregar_tarea()
    elif opcion == "2":
        tarea = input("Ingrese la tarea a marcar como completada: ")
        GestorTareas.marcar_completada(tarea)
    elif opcion == "3":
        GestorTareas.listar_tareas_pendientes()
    elif opcion == "4":
        GestorTareas.listar_tareas_completadas()
    elif opcion == "5":
        break
    else:
        print("Opcion incorrecta")
```
